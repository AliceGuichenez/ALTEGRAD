```python
import numpy as np
import networkx as nx
from time import time

import GraphData as data
from walks import generate_walks
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# To make nice plot for Latex
%matplotlib notebook
plt.rcParams.update({'font.size': 13}) # for Latex
plt.rcParams['figure.figsize'] = [8, 5]
```

# Compute features

```python
idxs, edgelists = data.get_graphs(N_train = 1000, test = False)
feats = []
for i, edgelist in enumerate(edgelists):
    g = nx.read_edgelist(edgelist)
    feats.append([len(g),
                  nx.density(g),
                  max(nx.core_number(g).values()),
                  
                  ])

feats = pd.DataFrame(feats, index = idxs, columns=["size", "density", "n_core"])
feats.describe()
```
